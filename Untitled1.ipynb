## Preprocessing the data
We are using Keras ImageDataGenerator to create Image augmentation. Image augmentation is a technique of applying different transformations such as a slit, rotate, resize, shift, etc. to the original images which results in multiple transformed copies of the same image. Each copy, however, is different from the other in certain aspects depending on the augmentation techniques you apply like shifting, rotating, flipping, etc. Applying these small amounts of variations on the original image does not change its target class but only provides a new perspective of capturing the object in real life. And so, we use it is quite often for building deep learning models.

For augmentation, we will be using

* image rescale: Rescaling for normalization of data.
* shear range transformation: Shear angle in a counter-clockwise direction in degrees.
* brightness_range: For random brightness change.

```python
def datapreprocessing(main_dir,batch_size):
    from tensorflow.keras.preprocessing.image import ImageDataGenerator
    
    train_gen = ImageDataGenerator(rescale=1.0/255,
                                   shear_range=0.2,
                                   fill_mode='nearest',
                                   brightness_range=[0.4,1.5],
                                   width_shift_range=0.2, height_shift_range=0.2,
                                   validation_split=0.3          # Dividing the dataset (train-validation split)
                                  )

    train_generator = train_gen.flow_from_directory(
        directory=main_dir,
        target_size=(128,128),                 # resizing it to (128,128) pixel.
        batch_size=batch_size,
        color_mode="grayscale",
        shuffle=True,
        subset='training',
        class_mode='categorical')   # set as training data

    validation_generator = train_gen.flow_from_directory(
        directory=main_dir,
        batch_size=batch_size,
        class_mode='categorical',
        target_size=(128,128),                # resizing it to (128,128) pixel.
        color_mode="grayscale",
        shuffle=True,
        subset='validation') # set as validation data
    
    return train_generator,validation_generator
```

```python
traingen,valigen = datapreprocessing(dir,20)
```

## Building Model's Architecture
We have used two convolutional layers in the model. The convolutional layer systematically apply learned filters to input images to create feature maps that summarize the presence of those features in the input. After the conv2d layer, we have added a pooling layer with a stride of 2 and a pooling size of 2 which operates upon each feature map separately to create a new set of the same number of pooled feature maps. This is done so that the limitation of con2d to record the precise position of features in the input is overcome.  
  
* layer1: The first layer is a 2D convolutional layer with 8 filters having the kernel size of 3 X 3, strides of 1, and having activation function as relu (Rectified Linear Unit). The input shape for this layer is the shape of the image which is 128 x 128.
* MaxPooling2D layer: This layer is a max-pooling 2dlayer added after conv2d layers with a stride of 2 and pooling size of 2.
* layer2: This layer is also a 2D convolutional layer with the same hyperparameters as that of layer1.
* MaxPooling2D layer: This layer is again a max-pooling 2dlayer added after second conv2d layers with a stride of 2 and pooling size of 2.
* Flatten layer: We have added a flattening layer to convert the data into a 1-dimensional array for inputting it to the next dense hidden layer. We flatten the output of the second max-pooling layer to create a single long feature vector.
* layer5: This is the first dense layer with 256 nodes with relu activation function.
* layer6: This is the second dense layer with 128 nodes with relu activation function.
* output: This is the output layer having the possibility of each of the three classes.

```python
def imageclf2(input_shape):
    from tensorflow import keras as ks

    model = ks.models.Sequential()
    model.add(ks.layers.Conv2D(8,(3,3),
                               strides=1,
                               activation="relu",
                               padding='same',
                               name="layer1",
                               input_shape=input_shape))
    model.add(ks.layers.MaxPooling2D(pool_size=2,strides=2))
    
    model.add(ks.layers.Conv2D(8,(3,3),strides=1,padding="same",activation="relu",name="layer2"))
    model.add(ks.layers.MaxPooling2D(pool_size=2,strides=2))
    
    
    model.add(ks.layers.Flatten())
    model.add(ks.layers.Dense(256,activation="relu",
                              name="layer5"))
    model.add(ks.layers.Dense(128,activation="relu",
                              name="layer6"))
    
    model.add(ks.layers.Dense(5,activation='softmax',
                              name="output"))
    model.summary()
    return model
```

## Building the Compiler
We have added a callback early stopping to prevent the overfitting of the model with the training data. For monitoring, we have added validation loss and patience of 4 which means that model will continue to train for 4 more epochs since the validation has started increasing.

For compiling the model we have added a categorical cross-entropy and adam optimizer with a learning rate of 0.001. We will be using the train and validation data that we have generated using the generator.

Along with this we also have plotted out the training and validation accuracy using matplotlib.

```python
def compiler2(model,train_generator,valid_generator,epchs,bsize=20,lr=0.0001):

    from tensorflow import keras as ks
    callbck = ks.callbacks.EarlyStopping(monitor='val_loss',patience=4,verbose=2,
                                         restore_best_weights=True,) 
    
    opt = ks.optimizers.Adam(learning_rate=lr)
    
    model.compile(loss='categorical_crossentropy',
                      optimizer=opt,
                      metrics=["accuracy"])
    history = model.fit(train_generator,
                        epochs=epchs,
                        callbacks=[callbck],
                        validation_data=valid_generator,
                        verbose = 1)

    return model,history
```

## Summary of the model
